In this regression task we will predict the percentage of marks using Python Scikit-Learn library that a student is expected to score based upon the number of hours they studied.

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing data
data=pd.read_csv('http://bit.ly/w-data')
print('data imported successfully')
data.head(10)
```

after,created the plot using dataset to find interrelation

```python
#plotting the  distribution of scores
from matplotlib import style
style.use('ggplot')
data.plot(x='Hours',y='Scores',style='o')
plt.title('hours vs scores')
plt.xlabel('hours studied')
plt.ylabel('percentage score')
plt.show()
```

From the above graph, you can clearly see that there is positive linear relation between the number of hours and percentage of score

**Preaparing the data**.

in this, I divided the data into "attributes" (inputs) and "labels" (outputs).

```python
X=data.iloc[:,:-1].values
Y=data.iloc[:,1].values
```

and then splited this data into training and test sets using Scikit-Learn's built-in train_test_split() method:

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,
                                                test_size=0.2,random_state=0)
```


**Training the Algorithm**


and then, trained the machine learning algorithm.

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
print('Training complete')
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line);
plt.show()
```

**Making Predictions**


Now that we have trained our algorithm, it's time to make some predictions.

```python
print(X_test)
y_pred=regressor.predict(X_test)
```

```python
#comparing actual vs predicted
df= pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})
df
```

```python
# You can also test with your own data
hours = [[9.25]]
own_pred = regressor.predict(hours)
print('no of hours={}'.format(hours))
print('predicted score ={}'.format(own_pred[0]))
```

**Evaluating the model**


Finally, let's evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square error. There are many such metrics.

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(Y_test, y_pred)) 
```
